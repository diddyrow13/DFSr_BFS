```python
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'C', 'D'],
    'C': ['A', 'B', 'D', 'E'],
    'D': ['B', 'C', 'E', 'F'],
    'E': ['C', 'D', 'F'],
    'F': ['D']
}
```

```python
visited = set()                         # set of nodes already visited


def DFSR(graph, node, visited):         # Graph recursive traversal
    print(node)                         # Print node since first time visiting
    visited.add(node)                   # Add node to visited set
    for adj in graph[node]:             # Check graph for new node from current
        if adj not in visited:          # Recurse if node not in visited (find new node)
            DFSR(graph, adj, visited)   # Found new node, visiting recursively


DFSR(graph, 'A', visited)
```

```python
def BFS(graph, node):                   # BFS
    queue = [node]                      # BFS is queue based. Add passed node to queue
    nodeList = set()                    # create set for already visited nodes
    nodeList.add(node)                  # Adds the first node that was passed
    while len(queue) > 0:               # Queue has to have an item
        curNode = queue.pop(0)          # Remove the current node from queue
        print(curNode)                  # Print current node since first time visiting
        nodes = graph[curNode]          # Get the adjacent nodes of the current node
        for adj in nodes:               # Check each adjacent node of current node
            if adj not in nodeList:     # If the node hasn't been seen...
                queue.append(adj)       # Add node to queue to visit later
                nodeList.add(adj)       # Add node to list since it has been seen


BFS(graph, 'A')
```
